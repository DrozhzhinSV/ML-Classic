# 1. Понятие функции многих переменных


$\underline{\text{Определение 1.}}$ $N$ - мерное координатное пространство - это множество всевозможных точек $\mathbf x = (x_{1}, x_{2}, \ldots, x_{N})$, каждая из которых определяется с помощью $N$ координат: $x_{1}, x_{2}, \ldots, x_{N}$.


$N$ - мерное координатное пространство называется **евклидовым**, если между любыми двумя его точками $M_{1}$ и $M_{2}$ определено расстояние $\rho(M_{1}, M_{2})$: 

$$
\rho(M_{1}, M_{2}) = \sqrt{(x_{1}^{'} - x_{1}^{''})^{2} + (x_{2}^{'} - x_{2}^{''})^{2} + \ldots + (x_{N}^{'} - x_{N}^{''})^{2}},
$$

которое удовлетворяет следующим условиям:

- **Симметричность**

$$
\rho(M_{1}, M_{2}) = \rho(M_{2}, M_{1}) \quad \forall M_{1}, M_{2};
$$

- **Положительная определенность**

$$
\rho(M_{1}, M_{2}) > 0 \quad \forall M_{1} \neq M_{2} \text{ и } \rho(M_{1}, M_{2}) = 0 \quad \forall M_{1} = M_{2};
$$

- **Неравенство треугольника**

$$
\rho(M_{1}, M_{3}) \le \rho(M_{1}, M_{2}) + \rho(M_{2}, M_{3}) \quad \forall M_{1}, M_{2}, M_{3}.
$$


Докажем, что введенная нами функция расстояния удовлетворяет последнему условию:

$$
\begin{multline}
\Big(\rho(M_{1}, M_{3})\Big)^{2} = \sum\limits_{i = 1}^{n}\big(x_{i}^{'} - x_{i}^{'''}\big)^{2} = \sum\limits_{i = 1}^{n}\Big(\big(x_{i}^{'} - x_{i}^{''}\big) + \big(x_{i}^{''} - x_{i}^{'''}\big)\Big)^{2} = \\ = \sum\limits_{i = 1}^{n}\big(x_{i}^{'} - x_{i}^{''}\big)^{2} +2 \sum\limits_{i = 1}^{n}\big(x_{i}^{'} - x_{i}^{''}\big)\big(x_{i}^{''} - x_{i}^{'''}\big) + \sum\limits_{i = 1}^{n}\big(x_{i}^{''} - x_{i}^{'''}\big)^{2} \le \Big\{\text{Нер-во Коши - Буняковского}\Big\} \le \\ \le \underbrace{\sum\limits_{i = 1}^{n}\big(x_{i}^{'} - x_{i}^{''}\big)^{2}}_{\Big(\rho(M_{1}, M_{2})\Big)^{2}} + \underbrace{2\sqrt{\sum\limits_{i = 1}^{n}\big(x_{i}^{'} - x_{i}^{''}\big)^{2}}\sqrt{\sum\limits_{i = 1}^{n}\big(x_{i}^{''} - x_{i}^{'''}\big)^{2}}}_{2\rho(M_{1}, M_{2})\rho(M_{2}, M_{3})} + \underbrace{\sum\limits_{i = 1}^{n}\big(x_{i}^{''} - x_{i}^{'''}\big)^{2}}_{\Big(\rho(M_{2}, M_{3})\Big)^{2}} \le \Big(\rho(M_{1}, M_{2}) + \rho(M_{2}, M_{3})\Big)^{2}
\end{multline}
$$

$\underline{\text{Примером}}$ евклидовых пространств являются числовая прямая $\mathbb{R}^{1}$, плоскость $\mathbb{R}^{2}$, пространство $\mathbb{R}^{3}$ и т.д.

$\underline{\text{Определение 2.}}$ Рассмотрим евклидово множество $\mathbf{X} \subset \mathbb{R}^{N}$. Если каждой точке ${\mathbf x} = (x_{1}, x_{2}, \ldots, x_{N}) \in \mathbf{X}$ ставится в соответствие по некоторому известному закону число $y$, причем единственное, то говорят, что на множестве $\mathbf{X}$ задана **функция** $y = f(x_{1}, x_{2}, \ldots, x_{N}) = f({\mathbf x})$. При этом множество $\mathbf{X}$ называется **областью определения** функции $y = f({\mathbf x})$, а множество $\mathbf{Y} = \Big\{y\Big| \exists {\mathbf x} \in \mathbf{X}: f({\mathbf x}) = y\Big\}$ - **множеством значений**. Элемент $y$ из множества значений функции $\mathbf{Y}$ называется **частным значением** функции в точке ${\mathbf x}$. 

$\underline{\text{Примеры:}}$

1. $$y = \sqrt{\sin{\big(x_{1}^{2} + x_{2}^{2}\big)}}$$

Найдем область определения данной функции:

$$
\sin{\big(x_{1}^{2} + x_{2}^{2}\big)} \ge 0 \quad \Rightarrow \quad \big(x_{1}^{2} + x_{2}^{2}\big) \in \big[2\pi k, \pi + 2\pi k\big], \quad k \in {\mathbb N} \cup {0}
$$

Поскольку функция синуса ограничена сверху единицей, то множество значений в данном случае будет отрезком $[0, 1]$.

Построим график области определения:

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(figsize = (8, 8), nrows = 1, ncols = 1)

alpha = np.linspace(0, 2 * np.pi, 200)

for i in range(1, 20):
    i = 20 - i
    radius = np.sqrt(np.pi * i)
    x1 = radius * np.cos(alpha)
    x2 = radius * np.sin(alpha)
    
    if (i % 2) == 1:
        axs.fill_between(x1, 0, x2, color = 'green', alpha = 0.7)
    else:
        axs.fill_between(x1, 0, x2, color = 'white')
        
    plt.plot(x1, x2, color = 'black')

axs.set_xlim((-np.sqrt(21 * np.pi), np.sqrt(21 * np.pi)))
axs.set_ylim((-np.sqrt(21 * np.pi), np.sqrt(21 * np.pi)))

axs.set_xlabel(r"$x_{1}$", fontsize = 15)
axs.set_ylabel(r"$x_{2}$", fontsize = 15)

axs.set_title(r"Визуализация области определения функции $y = \sqrt{\sin{(x_{1}^{2} + x_{2}^{2})}}$")
plt.show()
```

```python
import pandas as pd

# Зададим сетку
args = np.array(np.meshgrid(np.linspace(-4, 4, 200), np.linspace(-4, 4, 200))).reshape(2, -1)

# Определим функцию и рассчитаем ее значения по сетке
def f(x, y):
    return np.sqrt(np.sin(x**2 + y**2))
    
z = np.vectorize(f)(*args).astype(float)

# Удалим NaN
df = pd.DataFrame(dict(x = args[0], y = args[1], z = z)).dropna()


# Нарисуем
fig = plt.figure(figsize = (15, 10))


# Рисунок 1
ax = fig.add_subplot(2, 2, 1, projection = '3d')
ax.plot_trisurf(df['x'][(df.x <= 1.4) & (df.x >= -1.4) & (df.y <= 1.4) & (df.y >= -1.4)], 
                df['y'][(df.x <= 1.4) & (df.x >= -1.4) & (df.y <= 1.4) & (df.y >= -1.4)], 
                df['z'][(df.x <= 1.4) & (df.x >= -1.4) & (df.y <= 1.4) & (df.y >= -1.4)], 
                cmap = 'viridis', edgecolor = 'none')

ax.set_title(r"График функции $y = \sqrt{\sin{(x_{1}^{2} + x_{2}^{2})}}$ на множестве " +\
             r"$[-1.4; 1.4] \times [-1.4; 1.4]$", fontsize = 10)
ax.set_xlabel(r"$x_{1}$", fontsize = 10)
ax.set_ylabel(r"$x_{2}$", fontsize = 10)
ax.set_zlabel(r"$y$", fontsize = 10)

# Рисунок 2
ax = fig.add_subplot(2, 2, 2, projection = '3d')
ax.plot_trisurf(df['x'], df['y'], df['z'], cmap = 'viridis', edgecolor = 'none')

ax.set_title(r"График функции $y = \sqrt{\sin{(x_{1}^{2} + x_{2}^{2})}}$ на множестве " +\
             r"$[-4; 4] \times [-4, 4]$", fontsize = 10)
ax.set_xlabel(r"$x_{1}$", fontsize = 10)
ax.set_ylabel(r"$x_{2}$", fontsize = 10)
ax.set_zlabel(r"$y$", fontsize = 10)



# Рисунок 3
ax = fig.add_subplot(2, 2, 3, projection = '3d')
ax.plot_trisurf(df['x'][df.y >= 0], df['y'][df.y >= 0], df['z'][df.y >= 0], cmap = 'viridis', edgecolor = 'none')

ax.set_title(r"График функции $y = \sqrt{\sin{(x_{1}^{2} + x_{2}^{2})}}$ на множестве " +\
             r"$[-4; 4] \times [0; 4]$", fontsize = 10)
ax.set_xlabel(r"$x_{1}$", fontsize = 10)
ax.set_ylabel(r"$x_{2}$", fontsize = 10)
ax.set_zlabel(r"$y$", fontsize = 10)


ax = fig.add_subplot(2, 2, 4)
ax.scatter(df['x'], df['y'], color = plt.cm.viridis(df['z']))
ax.set_title(r"График функции $y = \sqrt{\sin{(x_{1}^{2} + x_{2}^{2})}}$ на множестве " +\
             r"$[-4; 4] \times [-4; 4]$: проекция на плоскость $Ox_{1}x_{2}$", fontsize = 10)
ax.set_xlabel(r"$x_{1}$", fontsize = 10)
ax.set_ylabel(r"$x_{2}$", fontsize = 10)

plt.show()
```

# 2. Предел функции многих переменных

$\underline{\text{Определение 3.}}$ $\varepsilon$ - **окрестностью** точки $M_{0} = \big(x_{1}^{0}, x_{2}^{0}, \ldots, x_{N}^{0}\big)$ будем называть $N$ - **мерный открытый шар** радиуса $\varepsilon$ с центром в точке $M_{0}$, т.е. множество точек $\big(x_{1}, x_{2}, \ldots, x_{N}\big)$:

$$
\big(x_{1} - x_{1}^{0}\big)^{2} + \big(x_{2} - x_{2}^{0}\big)^{2} + \ldots + \big(x_{N} - x_{N}^{0}\big)^{2} < \varepsilon^{2}. 
$$

Рассмотрим функцию $y = f\big(x_{1}, x_{2}, \ldots, x_{N}\big) = f(\mathbf x)$, определенную на некотором множестве $\mathbf{X} \subset \mathbb{R}^{N}$ и точку $A \in \mathbb R^{N}$, которая может не принадлежать множеству $\mathbf{X}$, но в любой ее проколотой $\varepsilon$ - окрестности имеются точки из множества $\mathbf{X}$. 


$\underline{\text{Определение 4.}}$ Число $b$ называется **пределом** функции $y = f(\mathbf x)$ в точке $A$, если для любого положительного числа $\varepsilon$ найдется положительное число $\delta$ такое, что для всех точек $M$ из области определения функции, удовлетворяющих условию

$$
0 < \rho\big(M, A\big) < \delta,
$$

выполняется условие

$$
\big|f(M) - b\big| < \varepsilon.
$$

То есть, если число $b$ является пределом функции $y = f(\mathbf x)$ в точке $A$, то значение функции в любой точке $M$, лежащей от точки $A$ на расстоянии не превышающем $\delta$, будет заключено между значениями $b - \varepsilon$ и $b + \varepsilon$. Т.о., мы можем сделать значение $f(\mathbf x)$ сколь угодно близким к $b$, выбирая точку $M$ достаточно близкой к точке $A$.


$\underline{\text{Отличие от функции одной переменной:}}$ основное отличие данных пределов заключается в том, что при нахождении предела функции одной переменной, необходимо исследовать как изменяется значение функции при стремлении к рассматриваемой точке по двум направлением: при движении к данной точке слева по кривой и справа. Соответственно для существования предела необходимо было равенство левого и правого предела. Если же мы имеем дело с функцией нескольких перменных, то таких направлений бесконечно много и для существования предела необходимо, чтобы пределы по всем направлениям существовали и были равны.  


$\underline{\text{Примеры:}}$

1. Покажем, что $\lim_{x\to 0, y \to 0}(x + y)\sin{\Big(\frac{1}{x}\Big)}\sin{\Big(\frac{1}{y}\Big)} = 0$:

В соответствии с определением, нам нужно показать, что для любой точки $(x, y)$ такой, что

$$
0 < \sqrt{(x - 0)^{2} + (y - 0)^{2}} = \sqrt{x^{2} + y^{2}} < \delta,
$$

выполняется условие

$$
\big|f(x, y) - 0\big| = \Bigg|(x + y)\sin{\Big(\frac{1}{x}\Big)}\sin{\Big(\frac{1}{y}\Big)}\Bigg| < \varepsilon.
$$

Из первого условия следует, что 

$$x^{2} + y^{2} < \delta^{2}.$$ 

Следовательно, 

$$x^{2} < \delta^{2} \text{ и } y^{2} < \delta^{2}.$$

Поскольку $\delta > 0$, то

$$
-\delta < x < \delta, \quad -\delta < y < \delta \quad \Rightarrow \quad |x + y| < 2\delta.
$$

Учитывая это получим

$$
\Bigg|(x + y)\sin{\Big(\frac{1}{x}\Big)}\sin{\Big(\frac{1}{y}\Big)}\Bigg| = \Bigg|(x + y)\Bigg|\cdot\Bigg|\sin{\Big(\frac{1}{x}\Big)}\Bigg|\cdot\Bigg|\sin{\Big(\frac{1}{y}\Big)}\Bigg| \le \Bigg|(x + y)\Bigg| < 2\delta
$$

Таким образом, достаточно взять $\delta = \frac{\varepsilon}{2}$.


2. Покажем, что у функции $f(x, y) = \frac{xy}{x^{2} + y^{2}}$ не существует предела в точке $(0, 0)$. Устремим точку $(x, y)$ к началу координат по прямой $y = kx$:

$$
\lim_{y = kx, x \to 0}f(x, y) = \lim_{x \to 0}\frac{kx^{2}}{x^{2}(1 + k^{2})} = \frac{k}{1 + k^{2}}.
$$

То есть, при движении точки $(x, y)$ к началу координат по различным прямым, получаются разные значения предела. Следовательно, предела в рассматриваемой точке не существует.


3. Покажем, что у функции $f(x, y) = \frac{x^{2}y}{x^{4} + y^{2}}$ не существует предела в точке $(0, 0)$. Устремим точку $(x, y)$ к началу координат по прямой $y = kx$:

$$
\lim_{y = kx, x \to 0}f(x, y) = \lim_{x \to 0}\frac{kx^{3}}{x^{2}(x^{2} + k^{2})} = \lim_{x \to 0}\frac{kx}{x^{2} + k^{2}} = 0.
$$

Аналогично, если устремим точку $(x, y)$ к началу координат по прямой $x = ky$:

$$
\lim_{x = ky, y \to 0}f(x, y) = \lim_{y \to 0}\frac{k^{2}y^{3}}{y^{2}(k^{4}y^{2} + 1)} = \lim_{y \to 0}\frac{k^{2}y}{k^{4}y^{2} + 1} = 0.
$$

То есть, при движении точки $(x, y)$ к началу координат по различным прямым, получается предел равный 0. Но, если мы устремим точку $(x, y)$ к началу координат по параболе $y = kx^{2}$, то получим следующее:

$$
\lim_{y = kx^{2}, x \to 0}f(x, y) = \lim_{x \to 0}\frac{kx^{4}}{x^{4}(1 + k^{2})} = \frac{k}{1 + k^{2}}.
$$

Таким образом, при движении точки $(x, y)$ к началу координат по различным параболам, получаются разные значения предела. Следовательно, предела в рассматриваемой точке не существует.

# 3. Непрерывность функции

Рассмотрим функцию $f(\mathbf x)$, определенную в некоторой области $\mathbf X \subset \mathbb R^{N}$. Пусть точка $M \in \mathbf X$ и, кроме того, в любой выколотой $\varepsilon$ окрестности точки $M$ содержатся точки из области определения рассматриваемой функции. 

$\underline{\text{Определение 5.}}$ Функция $y = f(\mathbf x)$ называется **непрерывной** в точке $M$, если у нее существует предел в этой точке, равный частному значению $f(M)$:

$$
\lim_{\mathbf x \to M} f(\mathbf x)= f(M). 
$$

Если функция $f(\mathbf x)$ непрерывна в каждой точке множества $\mathbf X$, то она непрерывна на этом множестве.

Рассмотрим произвольную точку $\tilde{\mathbf x}$ из области определения функции $y = f(\mathbf x)$. Пусть точка $M$ имеет координаты $(x_{1}, x_{2}, \ldots, x_{N})$. Выразим координаты точки $\tilde{\mathbf x}$ через координаты точки $M$: $\tilde{x}_{i} = x_{i} + \Delta x_{i}$. Величину 

$$
\Delta y = f(x_{1} + \Delta x_{1}, x_{2} + \Delta x_{2}, \ldots, x_{N} + \Delta x_{N}) - f(x_{1}, x_{2}, \ldots, x_{N}).
$$

назовем **приращением** функции $y = f(\mathbf x)$ в точке $M$.

Используя полученное соотношение, можем переформулировать условие непрерывности функции $y = f(\mathbf x)$ в точке $M$:

$$
\lim_{\Delta x_{1} \to 0, \ldots, \Delta x_{N} \to 0} \Delta y = 0.
$$

Это означает, что если функция $f(\mathbf x)$ непрерывна в точке $M$, то малым приращениям аргумента соотвутствуют малые изменения значений функции.



Также для функции многих переменных можно ввести понятие непрерывности по одной из переменных при фиксированном значении остальных. Для этого необходимо перейти к формуле **частного приращения** по переменной $x_{i}$:

$$
\Delta_{x_{i}} y = f(x_{1}, \ldots, x_{i - 1}, x_{i} + \Delta x_{i}, x_{i + 1}, \ldots, x_{N}) - f(x_{1}, x_{2}, \ldots, x_{N}).
$$

$\underline{\text{Определение 6.}}$ Функция $y = f(\mathbf x)$ будет непрерывной точке $M(x_{1}, x_{2}, \ldots, x_{N})$ по переменной $x_{i}$, если при фиксированных значениях остальных переменных выполнено равенство

$$
\lim_{\Delta x_{i} \to 0} \Delta_{x_{i}} y = 0.
$$

В данном случае, мы можем рассматривать функцию $f(\mathbf x)$ как функцию одной переменной $x_{i}$. Т.е., непрерывность функции по переменной $x_{i}$ означает непрерывность функции одной переменной.

$\underline{\text{Утверждение 1:}}$ Если функция $y = f(\mathbf x)$ непрерывна по совокупности переменных $(x_{1}, x_{2}, \ldots, x_{N})$ в точке $M(x_{1}, x_{2}, \ldots, x_{N})$, то она непрерывна в этой точке по каждой из переменных $x_{i}$. **Обратное неверно**. 

$\underline{\text{Пример:}}$

Рассмотрим функцию

$$
f(x, y) = \begin{cases}
\frac{xy}{x^{2} + y^{2}}&,& \text{ если } x^{2} + y^{2} \neq 0, \\
0&,& \text{иначе}.
\end{cases}
$$

Исследуем ее непрерывность по точке $(0, 0)$. Ранее уже было показано, что предел в этой точке не существует. Следовательно, функция не является непрерывной в данной точке по совокупности аргументов. 

Рассмотрим непрерывность по каждой из перменных в отдельности. Поскольку мы исследуем непрерывность в точке $(0, 0)$, то для расчета предела, например, по переменной $x$, нам нужно зафиксировать значение $y = 0$. В этом случае, $f(x, y) = f(x, 0) = 0$. Соответственно и предел в точке $(0, 0)$ равен $0$. Аналогично и для непрерывности по переменной $y$. 

# 4. Частные производные и дифференцируемость функции нескольких переменных


### Частные производные

Пусть функция $f(\mathbf x)$ определена на множестве $\mathbf X \subset \mathbb R^{N}$ и точка $M(x_{1}, x_{2}, \ldots, x_{N})$ является внутренней точкой множества $\mathbf X$, т.е. найдется проколотая $\varepsilon$ - окрестность точки $M$, целиком содержащаяся во множестве $\mathbf X$.

$\underline{\text{Определение 6.}}$ Если существует предел отношения частного приращения $\Delta_{x_{k}}y$ функции в точке $M$ к соответствующему приращению $\Delta x_{k}$ аргумента $x_{k}$ при $\Delta x_{k} \to 0$

$$
\lim_{\Delta x_{k} \to 0}\frac{\Delta_{x_{k}}y}{\Delta x_{k}} = \lim_{\Delta x_{k} \to 0}\frac{f\big(x_{1}, \ldots, x_{k - 1}, x_{k} + \Delta x_{k}, x_{k + 1}, \ldots, x_{N}\big) - f\big(x_{1}, x_{2}, \ldots, x_{N}\big)}{\Delta x_{k}},
$$

то этот предел называется **частной производной** функции $y = f(\mathbf x)$ в точке $M$ по аргументу $x_{k}$.

Частная производная функции по аргументу $x_{k}$ представляет собой обыкновенную производную функции одной переменной $x_{k}$ при фиксированных значениях остальных переменных.


$\underline{\text{Геометрический и физический смысл частной производной:}}$

**Физический смысл** частной производной функции $f(x_{1}, x_{2}, \ldots, x_{N})$ в точке $M$ по переменной $x_{k}$ - это скорость изменения функции в точке $М$ в направлении оси $Ox_{k}$.

**Геометрический смысл** рассмотрим на примере функции двух переменных $f(x, y)$. В данном случае, частная производная по переменной $f_{x}^{'}$ в точке $(x_{0}, y_{0})$ - это угловой коэффициент касательной $\Gamma_{x}$ к линии пересечения поверхности, заданной функцией $f(x, y)$ и плоскости $y = y_{0}$: $f_{x}^{'} = tg(\alpha)$. Аналогично для частной производной по $y$: $f_{y}^{'} = tg(\beta)$.   

<img src="images/geom_partial.png">


$\underline{\text{Замечания:}}$ 

1. Из существования у функции в точке всех частных производных не следует непрерывность: 

$$
f(x, y) = \begin{cases}
\frac{xy}{x^{2} + y^{2}}&,& \text{ если } x^{2} + y^{2} \neq 0, \\
0&,& \text{иначе}.
\end{cases}
$$

Ранее было показано, что данная функция не является непрерывной в точке $(0, 0)$. Однако, данная функция имеет в точке $(0, 0)$ частные производные по $x$ и $y$. Действительно, поскольку $f(x, 0) = 0$ и $f(0, y) = 0$, то 

$$
\frac{\partial f}{\partial x}\Big|_{(0, 0)} = 0, \quad \frac{\partial f}{\partial y}\Big|_{(0, 0)} = 0.
$$


2. Данное определение частной призводной может быть непригодным для граничных точек, т.к., например, в них не всегда можно вычислить значения частных приращений. Поэтому, если частная производная существует во внутренних точках, то обычно полагают

$$
\frac{\partial f}{\partial x_{k}}(M_{0}) = \lim_{M \to M_{0}}\frac{\partial f}{\partial x_{k}}(M)
$$

<img src="images/not_partial_dev.png">


$\underline{\text{Пример:}}$ Рассмотрим функцию

$$
f = \big|x - y\big|.
$$

1. Покажем, что в точках $x = y$ частных производных не существует

$$
\lim_{\Delta x \to 0 - 0}\frac{f(x + \Delta x, x) - f(x, x)}{\Delta x} = \lim_{\Delta x \to 0 - 0}\frac{\big|x + \Delta x - x\big| - \big|x - x\big|}{\Delta x} = \lim_{\Delta x \to 0 - 0}\frac{\big|\Delta x\big|}{\Delta x} = -1, \\
\lim_{\Delta x \to 0 + 0}\frac{f(x + \Delta x, x) - f(x, x)}{\Delta x} = \lim_{\Delta x \to 0 + 0}\frac{\big|x + \Delta x - x\big| - \big|x - x\big|}{\Delta x} = \lim_{\Delta x \to 0 + 0}\frac{\big|\Delta x\big|}{\Delta x} = 1.
$$

Следовательно, частной производной по переменной $x$ не существует. 

2. Покажем, что в точках $x = -y, x \neq 0$ частные производные существуют

$$
\lim_{\Delta x \to 0}\frac{f(x + \Delta x, -x) - f(x, -x)}{\Delta x} = \lim_{\Delta x \to 0}\frac{\big|x + \Delta x + x\big| - \big|x + x\big|}{\Delta x} = \lim_{\Delta x \to 0}\frac{\big|2x + \Delta x\big| - \big|2x\big|}{\Delta x} = \begin{cases}
\lim_{\Delta x \to 0}\frac{-\Delta x}{\Delta x} = -1&,& \text{ если } x < 0,\\
\lim_{\Delta x \to 0}\frac{\Delta x}{\Delta x} = 1&,& \text{ если } x > 0.
\end{cases}
$$

### Дифференцируемость

Вспомним ранее определенную формулу для приращения функции $y = f(\mathbf x)$ в точке $M(x_{1}, x_{2}, \ldots, x_{N})$, соответствующего приращениям аргументов $\Delta x_{1}, \Delta x_{2}, \ldots, \Delta x_{N}$:

$$
\Delta y = f(x_{1} + \Delta x_{1}, x_{2} + \Delta x_{2}, \ldots, x_{N} + \Delta x_{N}) - f(x_{1}, x_{2}, \ldots, x_{N}).
$$

$\underline{\text{Определение 7.}}$ Функция $y = f(\mathbf x)$ называется **дифференцируемой** в точке $M(x_{1}, x_{2}, \ldots, x_{N})$, если ее приращение в этой точке может быть представлено в виде

$$
\Delta y = \sum\limits_{i = 1}^{N}A_{i}\Delta x_{i} + \sum\limits_{i = 1}^{N}\alpha_{i}\big(\Delta x_{i}\big)\Delta x_{i}
$$

где $A_{i}$ - некоторые числа, не зависящие от $\Delta x_{i}$, а $\alpha_{i}$ - бесконечно малые функции при $\Delta x_{1} \to 0, \Delta x_{2} \to 0, \ldots, \Delta x_{N} \to 0$: 

$$
\lim_{\Delta x_{i} \to 0}\alpha_{i}\big(\Delta x_{i}\big) = 0,
$$

равные $0$ при $\Delta x_{1} = \Delta x_{2}, \ldots, \Delta x_{N} = 0$.

Таким образом, функция дифференцируема в точке $M$, если найдется окрестность этой точки, где ее приращение может быть представлено линейной функцией от приращений аргументов $\Delta x_{i}, \forall i = \overline{1, N}$.

$\underline{\text{Утверждение 2.}}$ Если функция дифференцируема в точке, то она непрерывна в этой точке. **Обратное неверно.**

Покажем это. Пусть $y = f(\mathbf x)$ дифференцируема в точке $M_{0}$, следовательно, ее приращение можно представить в виде:

$$
\Delta y(M_{0}) = \sum\limits_{i = 1}^{N}A_{i}\Delta x_{i} + \sum\limits_{i = 1}^{N}\alpha_{i}\big(\Delta x_{i}\big)\Delta x_{i}.
$$

Тогда 

$$
\lim_{\Delta x_{1} \to 0, \ldots, \Delta x_{N} \to 0}\Delta y(M_{0}) = \lim_{\Delta x_{1} \to 0, \ldots, \Delta x_{N} \to 0}\Bigg(\sum\limits_{i = 1}^{N}A_{i}\Delta x_{i} + \sum\limits_{i = 1}^{N}\alpha_{i}\big(\Delta x_{i}\big)\Delta x_{i}\Bigg) = 0.
$$

Таким образом, функция $f(\mathbf x)$ непрерывна в точке $M_{0}$.


$\underline{\text{Пример:}}$ Функция 

$$
y = f(x_{1}, x_{2}, \ldots, x_{N)} = \Bigg|\sum\limits_{i = 1}^{N}x_{i}\Bigg|
$$

является непрерывной в точке $(0, 0, \ldots, 0)$:

$$
\lim_{\Delta x_{1} \to 0, \ldots, \Delta x_{n} \to 0}\Delta y(0, 0, \ldots, 0) = \lim_{\Delta x_{1} \to 0, \ldots, \Delta x_{n} \to 0}\Bigg|\sum\limits_{i = 1}^{N}\Delta x_{i}\Bigg| = 0.
$$

При этом, величина

$$
\Delta y(0, 0, \ldots, 0) = \Bigg|\sum\limits_{i = 1}^{N}\Delta x_{i}\Bigg|
$$

не является линейной по переменным $\Delta x_{1}, \ldots, \Delta x_{N}$, следовательно, не является дифференцируемой в этой точке.

$\underline{\text{Утверждение 3.}}$ Если функция $y = f(x_{1}, x_{2}, \ldots, x_{N})$ дифференцируема в точке $M(x_{1}, x_{2}, \ldots, x_{N})$, то в этой точке существуют частные производные по всем аргументам, причем

$$
\frac{\partial y}{\partial x_{i}} = A_{i}.
$$

**Обратное неверно.**

Действительно, так как функция $y = f(\mathbf x)$ дифференцируема, то частное приращение этой функции по переменной $x_{i}$ в точке $M$

$$
\Delta_{x_{i}}y = A_{i}\Delta x_{i} + \alpha_{i}\Delta x_{i}.
$$

Тогда

$$
\frac{\partial y}{\partial x_{i}} = \lim_{\Delta x_{i} \to 0}\frac{\Delta_{x_{i}}y}{\Delta_{x_{i}}} = \lim_{\Delta x_{i} \to 0}\frac{A_{i}\Delta x_{i} + \alpha_{i}\Delta x_{i}}{\Delta_{x_{i}}} = \lim_{\Delta x_{i} \to 0}\big(A_{i} + \alpha_{i}\big) = A_{i}
$$

$\underline{\text{Пример:}}$ Рассмотрим функцию 

$$
f(x, y) = \begin{cases}
1, \text{ если } x = 0 \text{ или } y = 0,\\
0, \text{ иначе.}
\end{cases}
$$

Найдем частное приращение по переменной $x$ в точке $(0, 0)$:

$$
\Delta_{x}f(0, 0) = f(\Delta x, 0) - f(0, 0) = 1 - 1 = 0 \Rightarrow \lim_{\Delta x \to 0}\frac{\Delta_{x}f}{\Delta_{x}} = 0,
$$

следовательно, $\frac{\partial f}{\partial x}(0, 0) = 0$. Аналогично можно показать, что $\frac{\partial f}{\partial y}(0, 0) = 0$.

Но функция $f(x, y)$ не является непрерывной в точке $(0, 0)$, а потому не дифференцируема в этой точке.

# 5. Производная по направлению и градиент функции


### Понятие производной по направлению и ее связь с частными производными

Пусть функция $f(x_{1}, x_{2}, \ldots, x_{N})$ определена на множестве $\mathbf X \subset \mathbb R^{N}$. Рассмотрим точку $M_{0} \in \mathbf X$ такую, что рассматриваемая функция определена в некоторой ее окрестности.

Проведем через точку $M_{0}$ произвольную прямую $L$ и выберем на ней одно из двух возможных направлений, которое задается единичным вектором $\overrightarrow{l}$. Известно, что если единичный вектор составляет с осями координат углы $\alpha_{1}, \alpha_{2}, \ldots, \alpha_{N}$, то координаты этого вектора будут иметь вид: $\Big\{\cos{\alpha_{1}}, \cos{\alpha_{2}}, \ldots, \cos{\alpha_{N}}\Big\}$. 

<img src="images/dif_by_direct.png">

Отметим на прямой $L$ точку $M$ так, чтобы она лежала в указанной окрестности точки $M_{0}$ и введем величину направленного отрезка $\overrightarrow{MM_{0}}$:

$$
h = \begin{cases}
|\overrightarrow{MM_{0}}|, \text{ если направление } \overrightarrow{MM_{0}} \text{ совпадает с направление вектора } \overrightarrow{l},\\
- |\overrightarrow{MM_{0}}|, \text{ иначе}.
\end{cases}
$$


$\underline{\text{Определение 8.}}$ Если существует 

$$
\lim_{h \to 0}\frac{f(M) - f(M_{0})}{h} = \lim_{h \to 0}\frac{f(M_{0} + h\overrightarrow{l}) - f(M_{0})}{h},
$$

то он называется **производной** функции $f(x_{1}, x_{2}, \ldots, x_{N})$ в точке $M_{0}$ **по направлению** $\overrightarrow{l}$ и обозначается $\frac{\partial f}{\partial l}(M_{0})$. 


Если функция дифференцируема в точке $M_{0}$, то ее производная по направлению в этой точке может быть выражена через частные производные следующим образом:

$$
\frac{\partial f}{\partial l}(M_{0}) = \frac{\partial f}{\partial x_{1}}(M_{0})\cos(\alpha_{1}) + \frac{\partial f}{\partial x_{2}}(M_{0})\cos(\alpha_{2}) + \ldots + \frac{\partial f}{\partial x_{N}}(M_{0})\cos(\alpha_{N}).
$$

 
$\underline{\text{Физический смысл:}}$ если функция дифференцируема в точке $M_{0}$, то в этой точке скорость изменения функции по заданному направлению $\overrightarrow{l}$ является линейной комбинацией скоростей изменения этой функции по направлениям координатных осей, причем коэффициентами этой линейной комбинации являются координаты единичного вектора $\overrightarrow{l}$. Данные коэффициентами являются весами, показывая какой вклад вносит каждая частная производная. 


$\underline{\text{Пример:}}$ Рассмотрим функцию 

$$
f(x, y) = |x - y|.
$$

Ранее уже было показано, что она не имеет частных производных во всех точках $x = y$. В терминах производной по направлению это означает, что у рассматриваемой функции в точках $x = y$ не существует производных в направлениях $\overrightarrow{l} = (1, 0)$ и $\overrightarrow{l} = (0, 1)$, т.е. в направлении осей $Ox$ и $Oy$. Исследуем существование производных в данных точках по другим направлениям.

<img src="images/der_by_dir_1.png">

Пусть направление задается единичным вектором с координатами $\frac{1}{\sqrt{\alpha^{2} + \beta^{2}}}\big(\alpha, \beta\big), \beta, \alpha \in \mathbb R$ и $\alpha$ и $\beta$ не равны нулю одновременно. Тогда

$$
\lim_{h \to 0}\frac{f\Bigg(x + h\frac{\alpha}{\sqrt{\alpha^{2} + \beta^{2}}}, x + h\frac{\beta}{\sqrt{\alpha^{2} + \beta^{2}}}\Bigg) - f(x, x)}{h} = \lim_{h \to 0}\frac{\Bigg|x + \frac{h\alpha}{\sqrt{\alpha^{2} + \beta^{2}}} - x - \frac{h\beta}{\sqrt{\alpha^{2} + \beta^{2}}}\Bigg|}{h} = \lim_{h \to 0}\frac{\Bigg|\frac{h\alpha}{\sqrt{\alpha^{2} + \beta^{2}}} - \frac{h\beta}{\sqrt{\alpha^{2} + \beta^{2}}}\Bigg|}{h} = \Bigg|\frac{\alpha - \beta}{\sqrt{\alpha^{2} + \beta^{2}}}\Bigg|\lim_{h \to 0}\frac{|h|}{h}.
$$

Данный предел существует только при $\alpha = \beta$, поскольку иначе пределы слева и справа не будут совпадать. Условие $\alpha = \beta$ определяет либо единичный вектор $\Big(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\Big)$, либо единичный вектор $\Big(-\frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}}\Big)$, который задает направление вдоль оси $x = y$.


<img src="images/der_by_dir_2.png">

### Градиент функции

$\underline{\text{Определение 9.}}$ **Градиентом** дифференцированной функции $f(x_{1}, x_{2}, \ldots, x_{N})$ в точке $M_{0}$ называется вектор, координаты которого определяются частными производними функции $f$ в точке $M_{0}$:

$$
grad f(M_{0}) = \Bigg\{\frac{\partial f}{\partial x_{1}}\big(M_{0}\big), \frac{\partial f}{\partial x_{2}}\big(M_{0}\big), \ldots, \frac{\partial f}{\partial x_{N}}\big(M_{0}\big)\Bigg\}.
$$

Таким образом производную по направлению можно переписать следующим образом:

$$
\frac{\partial f}{\partial l}(M_{0}) = \Big(\overrightarrow{l}, grad f(M_{0})\Big) = \frac{\partial f}{\partial x_{1}}(M_{0})\cos(\alpha_{1}) + \frac{\partial f}{\partial x_{2}}(M_{0})\cos(\alpha_{2}) + \ldots + \frac{\partial f}{\partial x_{N}}(M_{0})\cos(\alpha_{N}).
$$

Покажем, что градиент функциив точке $M_{0}$ характеризует направление и величину максимального роста этой функции в точке $M_{0}$. Для этого необходимо убедиться в том, что производная функции $f$ в точке $M_{0}$ по направлению, определяемому градиентом этой функции в указанной точке, имеет максимальное значение по сравнению с производной по любому другому направлению в точке $M_{0}$, а значение указанной производной равно длине $|grad f|$.

Поскольку скалярное произведение двух векторов равно произведению длин этих векторов на косинус угла между ними, то 

$$
\frac{\partial f}{\partial l}(M_{0}) = \big|\overrightarrow{l}\big| \cdot |grad f| \cdot \cos{\psi},
$$

где $\psi$ - угол между векторами $\overrightarrow{l}$ и $grad f$. Так как вектор $\overrightarrow{l}$ единичный, то $\big|\overrightarrow{l}\big| = 1$. Соответственно,

$$
\frac{\partial f}{\partial l}(M_{0}) = |grad f| \cdot \cos{\psi}.
$$

Из последней формулы получаем, что максимальное значение производной по направлению будет достигаться при $\cos{\psi} = 1$. А из этого следует, во - первых, что направление вектора $\overrightarrow{l}$ совпадает с направлением $grad f$, а, во - вторых, что максимальное значение производной по направлению равно длине вектора $grad f$.


# 6. Метод градиентного спуска

Понятие градиента играет ключевую роль в машинном обучении. При построении различных прогностических моделей решается задача оптимизации функции ошибок, то есть задача нахождения минимального значения функции. 

Как правило, для решения этой задачи используется метод градиентного спуска. Основная идея данного метода заключается в том, чтобы идти в направлении наискорейшего спуска, а это направление, как мы показали выше, определяется направлением антиградиента (мы показали, что градиент функции  в точке характеризует направление и величину максимального роста этой функции в точке. Соответственно вектор градиента взятый со знаком "минус", характеризует величину максимального убывания функции в точке).

В самом общем виде, алгоритм градиентного спуска выглядит следующим образом:

1. Инициализация начального значения $\mathbf x^{0}$. Можно выбирать случайно, но лучше делать осознанный выбор, иначе может возникнуть проблема, которая была описана при рассмотрении гиперболического параболоида: можем попасть в точку, градиент которой равен 0 и которая не является глобальным минимумом;


2. Рассчитать значение градиента функции в точке $\mathbf x^{t - 1}$, рассчитанной на предыдущей итерации;


3. Перерасчитать точки $\mathbf x^{t}$  по следующей формуле:

$$
\mathbf x^{t} = \mathbf x^{t - 1} - \eta_{t} grad(\mathbf x^{t - 1}), 
$$

где $\eta_{t}$ - скорость градиентного спуска. Данный параметр можно выбрать, например, постоянным, а можно, например, уменьшать с каждым шагом. В случае, если шаг маленький, скорость сходимости будет также маленькой. При большом шаге, есть риск "перепрыгнуть" через точку минимума.


4. Оценить разницу двух последних приближений:

$$
||\mathbf x^{t} - \mathbf x^{t - 1}|| \le \varepsilon,
$$

и в случае, если она превышает некоторое заранее заданное значение $\varepsilon$, перейти к шагу 2, в противном случае - завершить процесс.

Подробнее о различных алгоритмах метода градиентного спуска можно посмотерть, например, [здесь](https://vbystricky.github.io/2018/03/optimization_grad_desc.html).

Реализуем простейшую версию градиентного спуска. Для начала введем понятие формулы Тейлора:

$$
f(x_{0}) = f(a) + f^{'}(a)(x_{0} - a) + f^{''}(a)\frac{(x_{0} - a)^{2}}{2!} + \ldots + f^{(n)}(a)\frac{(x_{0} - a)^{n}}{n!} = \sum\limits_{k = 0}^{n}f^{(k)}(a)\frac{(x_{0} - a)^{k}}{k!}.
$$

Рассмотрим два случая

$$
\begin{cases}
x_{0} = x + h,\\
a = x.
\end{cases}, \quad
\begin{cases}
x_{0} = x - h,\\
a = x.
\end{cases}
$$

Тогда

$$
f(x + h) = f(x) + f^{'}(x)h + f^{''}(x)\frac{h^{2}}{2!} + f^{'''}(x)\frac{h^{3}}{3!} + \ldots + f^{(n)}(x)\frac{h^{n}}{n!},\\ 
f(x - h) = f(x) - f^{'}(x)h + f^{''}(x)\frac{h^{2}}{2!} - f^{'''}(x)\frac{h^{3}}{3!} + \ldots + f^{(n)}(x)\frac{(-h)^{n}}{n!}
$$

Следовательно, с одной стороны,

$$
f^{'}(x) = \frac{f(x + h) - f(x)}{h} - f^{''}(x)\frac{h}{2!} - f^{'''}(x)\frac{h^{2}}{3!} - \ldots - f^{(n)}(x)\frac{h^{n - 1}}{n!},
$$

а, с другой стороны, 

$$
f^{'}(x) = \frac{f(x) - f(x - h)}{h} + f^{''}(x)\frac{h}{2!} - f^{'''}(x)\frac{h^{2}}{3!} + \ldots - f^{(n)}(x)\frac{(-h)^{n - 1}}{n!}.
$$

Если сложить эти выражения, то члены с четными производными сократятся и мы получим

$$
f^{'}(x) = \frac{f(x + h) - f(x - h)}{2h} + o\big(h\big).
$$

Данную формулу будем использовать для численного вычисления частных производных при расчете градиента функции. **Вопрос:** для чего мы рассмотрели два разложения функции в ряд Тейлора, а не остановились на одном?  

```python
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib.animation import FuncAnimation

%matplotlib notebook
```

```python
h = 1e-8

#----------   Функция расчета значений частных производных 
#                      (используем формулу Тейлора)   ---------- 

# f - функция 
# p - точка
def calc_gradient(f, p):
    (x, y) = p
    return np.array([
        (f(x + h, y) - f(x - h, y)) * 0.5 / h,
        (f(x, y + h) - f(x, y - h)) * 0.5 / h])


#----------   Функция расчета градиентного спуска:   ---------- 

#   f - функция 
#   p - старотовая точка
# eta - шаг алгоритма
#   n - кол-во шагов
# eps - оценка расстояния между соседними точками

def calc_gradient_descent(f, p, eta, n, eps):
    for i in range(n):
        df = calc_gradient(f, p)              #считаем градиент
        dfu = df / np.sqrt((df * df).sum())
        yield (p, dfu)
        
        if (eta * sum(dfu**2) <= eps):
            break
            
        p = p - eta * dfu
            
```

```python
# Построение графика функции
def plot_surface(f, ax, n_points_per_axis = 100):
    x_range = f['x_range']
    y_range = f['y_range']
    
    xs, ys = np.meshgrid(
        np.linspace(x_range[0], x_range[1], n_points_per_axis),
        np.linspace(y_range[0], y_range[1], n_points_per_axis))
    
    zs = np.vectorize(f['f'])(xs, ys)
    
    ax.plot_surface(
        xs, ys, zs,
        cmap = cm.coolwarm,
        linewidth = 0,
        antialiased = False,
        alpha = 0.4)
    
    ax.view_init(*f['view_init'])
    ax.set_zlim(*f['z_lim'])
    
# Отрисовка пути, пройденного точкой по градиенту  
def update_grads(n, lines, grad_data):
    for ln, gd in zip(lines, grad_data):
        ln.set_data(gd[0 : 2, : n])
        ln.set_3d_properties(gd[2, : n])
    return lines[1]
```

```python
# Словарь функций:
a = 10

functions_dict = dict(
    
    #Гиперболический параболоид
    saddle_point = dict(
        f = lambda x, y: x * x - y * y,             # вид функции
        x_range   = (-a, a),                        # область определения по x 
        y_range   = (-a, a),                        # область определения по y 
        x_lim     = (-a - 1, a + 1),                # ограничения для отрисовки графика по оси x
        y_lim     = (-a - 1, a + 1),                # ограничения для отрисовки графика по оси y
        z_lim     = (-1.1 * a * a, 1.1 * a * a),    # ограничения для отрисовки графика по оси z
        view_init = (60, 60)),                      # настройки камеры
    
    #Эллиптический параболоид
    elliptic_paraboloid = dict(
        f = lambda x, y: x * x + y * y,             # вид функции
        x_range   = (-a, a),                        # область определения по x 
        y_range   = (-a, a),                        # область определения по y 
        x_lim     = (-a - 1, a + 1),                # ограничения для отрисовки графика по оси x
        y_lim     = (-a - 1, a + 1),                # ограничения для отрисовки графика по оси y
        z_lim     = (0, 2.1 * a * a),               # ограничения для отрисовки графика по оси z
        view_init = (60, 60)),                      # настройки камеры
    
    beale = dict(
        #f = lambda x, y: (1.5 - x + x * y)**2 + (2.25 - x + x * (y**2))**2 + (2.625 - x + x * (y**3))**2,
        f = lambda x, y: x**2 + y**2 - 2*x,
        x_range   = (-0.5 * a, 0.5 * a),
        y_range   = (-0.5 * a, 0.5 * a),
        x_lim     = (-0.55 * a, 0.55 * a),
        y_lim     = (-0.55 * a, 0.55 * a),
        z_lim     = (-10, 100),#0.5 * a**6),
        view_init = (30, 30))
)
```

```python
# Выбираем функцию
func = functions_dict['saddle_point']

# Определяем стартовые точки для метода градиентного спуска
start_points = [
    
    np.array([-5, 0]),
    np.array([5, -8]),
    np.array([-5, 1.5]),
    np.array([1, -5.0])
]

grad_data = []


eta = 0.005 # Шаг алгоритма
n = 2000   # Кол-во шагов
eps = 1e-8 # Оценка для расстояний между соседними шагами

# Для каждой стартовой точки рассчитываем последовательность точек, лежащих на пути градиентного спуска
for p0 in start_points:
    (ps, _) = zip(*list(calc_gradient_descent(func['f'], p0, eta, n, eps)))
    xs, ys = list(zip(*ps))
    zs = np.vectorize(func['f'])(xs, ys)
    grad_data.append(np.array([xs, ys, zs]))
```

```python
fig = plt.figure(figsize = (9, 5))
ax = Axes3D(fig)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


plot_surface(func, ax)
ax.scatter(*zip(*map(lambda gd: gd[:, 0], grad_data)), c = 'r', marker = 'o')



lines = []
color_points = ['red', 'green', 'black', 'magenta']
for i, gd in enumerate(grad_data):
    ln, = ax.plot(gd[0, :1], gd[1, :1], gd[2, :1], c = color_points[i], marker = None)
    lines.append(ln)
    
anim = FuncAnimation(fig, update_grads, len(zs), interval = 1, fargs = (lines, grad_data))
plt.show()
```
