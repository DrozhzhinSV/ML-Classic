# 1. Понятие определенного интеграла


### Определение и геометрический смысл:

Пусть функция $f(x)$ определена на отрезке $[a, b]$. Рассмотрим разбиение этого отрезка произвольными точками $x_{0}, x_{1}, \ldots, x_{n}$: 

$$
a = x_{0} < x_{1} < x_{2} < \ldots < x_{n -1} < x_{n} = b
$$ 

на $n$ частичных сегментов $[x_{0}, x_{1}], [x_{1}, x_{2}], \ldots, [x_{n - 1}, x_{n}]$. 

Данное разбиение будем обозначать через $T$. Через $\xi_{i}$ обозначим произвольную точку отрезка $[x_{i- 1}, x_{i}]$, а через $\Delta x_{i}$ - длину данного отрезка.

$\underline{\text{Определение 1.}}$ Величина

$$
I(x_{1}, \ldots, x_{n}, \xi_{1}, \ldots, \xi_{n}) = f(\xi_{1})\Delta x_{1} + f(\xi_{2})\Delta x_{2} + \ldots + f(\xi_{n})\Delta x_{n} = \sum\limits_{i = 1}^{n}f(\xi_{i})\Delta x_{i}, 
$$

называется **интегральной суммой** функции $f(x)$ при выбранном разбиении $T$ и множестве промежуточных точек $\xi_{i}$.


$\underline{\text{Геометрический смысл:}}$

Предположим, что функция $y = f(x)$ всюда неотрицательна на сегменте $[a, b]$ и непрерывна. Рассмотрим криволинейную трапецию, ограниченную графиком функции $y = f(x)$ на сегменте $[a, b]$, двумя прямыми $x = a$ и $x = b$, а также осью абсцисс. Введенная интегральная сумма представляет собой сумму площадей прямоугольников, ширина которых равна длине соответствующего сегмента $[x_{i - 1}, x_{i}]$, а высота - значению функции в промежуточной точке $\xi_{i}$. То есть, интегральная сумма - это некоторая оценка для площади рассматриваемой криволинейной трапеции. 

<img src="images/geom_int.png">

Геометрический смысл для знакопеременной функции можно ввести 2 способами:

- Во-первых, можно график исходной функции поднять на величину 

$$
\Big|\min_{x \in [a, b]}f(x)\Big|.
$$

В этом случае, мы получим неотрицательную функцию, интегральная сумма которой будет неотрицательна. Следовательно, ее можно ассоциировать с площадью. Чтобы перейти к исходной интеральной сумме, из площади полученной фигуры достаточно отнять площадь прямоугольника со сторонами $(b - a)$ и $\Big|\min_{x \in [a, b]}f(x)\Big|$.

<img src="images/geom_int_integral.png">


- Другой подход заключается в обобщении понятия площади - [**ориентированная площадь**](https://ru.wikipedia.org/wiki/%D0%9E%D1%80%D0%B8%D0%B5%D0%BD%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%BD%D0%B0%D1%8F_%D0%BF%D0%BB%D0%BE%D1%89%D0%B0%D0%B4%D1%8C). В этом случае, если фигура лежит под осью абсцисс, то ее площадь считают отрицательной.

 
Если функция не является непрерывной, то на каждом участке непрерывности считается своя интегральная сумма, а затем они суммируются.


Обозначим $\Delta = \max_{x_{i}}{\Delta x_{i}}$ - **диаметр разбиения**.


$\underline{\text{Определение 2.}}$ Если существует конечный предел интегральных сумм при стремлении к $0$ диаметра разбиения вне зависимости от выбора разбиения и выбора точке $\xi_{i}$, то этот предел называется **определенным интегралом** от функции $f(x)$ по сегменту $[a, b]$, а сама функция $f(x)$ называется **интегрируемой**.


$\underline{\text{На $\varepsilon - \delta$ языке это означает следующее:}}$ 

Тогда при $\Delta \to 0$ для любого $\varepsilon > 0$ найдется $\delta > 0$, что для любого разбиения $T$ сегмента $[a, b]$, диаметр разбиения которого меньше $\delta$, независимо от выбора точек $\xi_{i}$ на сегментах $[x_{i - 1}, x_{i}]$ значение интегральных сумм будет заключено между значениями $b - \varepsilon$ и $b + \varepsilon$, где $b$ - некоторое конечное значение.

### Классы интегрируемых функций:

$\underline{\text{Утверждение 1.}}$ Неограниченная на сегменте $[a, b]$ функция не являтеся интегрируемой на этом сегменте.

Действительно, в данном случае, при любом разбиении сегмента $[a, b]$ интегральная сумма может быть сделана сколько угодно большой. Из этого следует, что не будет существовать предела интегральных сумм при стремлении к 0 диаметра разбиения.

Таким образом, в дальнейшем имеет смысл рассматривать только ограниченные на сегменте $[a, b]$ функции. **Всякая ли ограниченная на сегменте $[a, b]$ функция будет интегрируемой?** 

$\underline{\text{Примеры:}}$

1. $$f(x) = c = const, \quad x \in [a, b].$$

В этом случае, какое бы разбиение $T$ мы не рассматривали, получим следующее:

$$
I = \sum\limits_{i = 1}^{n}f(\xi_{i})\Delta x_{i} = \sum\limits_{i = 1}^{n}c\Delta x_{i} = c\sum\limits_{i = 1}^{n}\Delta x_{i} = c(b - a) \quad \Rightarrow \quad \int_{a}^{b}cdx = c(b - a).
$$

Таким образом, постоянная на сегменте $[a, b]$ функция - интегрируема на этом сегменте.


2. Функция Дирихле

$$
f(x) = \begin{cases}
1, \text{ если } x \text{ - рациональное число,}\\ 
0, \text{ если } x \text{ - иррациональное число.}
\end{cases}
$$

В данном случае, значение интегральной суммы будет зависеть от выбора точек $\xi_{i}$ и будет колебаться от $0$, если все точки $\xi_{i}$ - иррациональны до $b - a$, если все точки $\xi_{i}$ - рациональны. Таким образом, предела интегральных сумм существовать не будет, а, следовательно, не будут существовать и интеграла. 

Существует необходимое и достаточное условие интегрируемости функции. Для начала вспомним, что такое верхняя и нижняя точные грани функции.

$\underline{\text{Определение 3.}}$ Число $M$ называется **точной верхней гранью функции** $f(x)$ на некотором множестве $\mathbf X$, если выполняются два условия:

   - $f(x) \le M \quad \forall x \in \mathbf X$
   
   - $\forall \varepsilon > 0 \quad \exists x\in \mathbf X: f(x) > M - \varepsilon.$
   
Аналогично вводится понятие **точной нижней грани функции** $f(x)$. 

Введем две величины, которые будем называть **верхней** и **нижней интегральными суммами**:

$$
S = \sum\limits_{i = 1}^{n}M_{i}\Delta x_{i}, \quad s = \sum\limits_{i = 1}^{n}m_{i}\Delta x_{i},
$$

где $M_{i}, m_{i}$ - точные верхние и нижние грани соответствующих частичных сегментов $[x_{i - 1}, x_{i}]$. 

<img src="images/geom_int2.png">

$\underline{\text{Утверждение 2. Необходимое и достаточное условие интегрируемости}}$  (Ильин, Позняк стр. 335)

Для того, чтобы ограниченная на сегменте $[a, b]$ функция $f(x)$ была интегрируема на этом сегменте, необходимо и достаточно, чтобы для любого $\varepsilon > 0$ нашлось такое разбиение этого сегмента $T$, что выполняется следующее неравенство:

$$
S - s \le \varepsilon.
$$


$\underline{\text{Классы интегрируемых функций:}}$

- Непрерывные функции (Ильин, Позняк стр. 341)

- Разрывные функции, имеющие конечное число точек разрыва (Ильин, Позняк стр. 342)

- Монотонные функции (Ильин, Позняк стр. 344)

### Основные свойства определенного интеграла:

1. $$\int\limits_{a}^{a}f(x)dx = 0$$


2. $$\int\limits_{a}^{b}f(x)dx = -\int\limits_{b}^{a}f(x)dx$$


Первые два свойства рассматриваются как соглашения, которые следуют из определения интеграла.

3. Если функции $f(x)$ и $g(x)$ интегрируемы на сегменте $[a, b]$, то функция $\alpha f(x) + \beta g(x)$ также интегрируема на этом сегменте и справедлива следующая формула 

$$
\int\limits_{a}^{b}\Big(\alpha f(x) + \beta g(x)\Big)dx = \alpha \int\limits_{a}^{b}f(x)dx + \beta \int\limits_{a}^{b}g(x)dx, \quad \alpha, \beta \in \mathbb R.
$$

При любом разбиении сегмента $[a, b]$ и любом выборе точек $\xi_{i}$ для интегральных сумм будет справедливо следующее соотношение:

$$
\sum\limits_{i = 1}^{n}\Big(\alpha f(\xi_{i}) + \beta g(\xi_{i})\Big)\Delta x_{i} = \alpha\sum\limits_{i = 1}^{n}f(\xi_{i})\Delta x_{i} + \beta\sum\limits_{i = 1}^{n}g(\xi_{i})\Delta x_{i}.
$$

Из начальных условий следует, что существует предел правой части. Следовательно, существует предел левой части и справедлива приведенная выше формула. 




4. Если функции $f(x)$ и $g(x)$ интегрируемы на сегменте $[a, b]$, то произведение этих функций также интегрируемо на этом сегменте.

Разобъем сегмент $[a, b]$ на $n$ частичных сегментов и введем следующие обозначения: 

- $M_{i}, m_{i}$ - точные верхняя и нижняя грани для произведения функций $f(x)$ и $g(x)$ на частичном сегменте $[x_{i - 1}, x_{i}]$   


- $M_{i}^{f}, m_{i}^{f}$ - точные верхняя и нижняя грани для функции $f(x)$ на частичном сегменте $[x_{i - 1}, x_{i}]$ 


- $M_{i}^{g}, m_{i}^{g}$ - точные верхняя и нижняя грани для функции $g(x)$ на частичном сегменте $[x_{i - 1}, x_{i}]$. 


Тогда 

$$
S - s = \sum\limits_{i = 1}^{n}(M_{i} - m_{i})\Delta x_{i} \le \sum\limits_{i = 1}^{n}\big(M_{i}^{f}M_{i}^{g} - m_{i}^{f}m_{i}^{g}\big)\Delta x_{i} = \sum\limits_{i = 1}^{n}\big(M_{i}^{f} - m_{i}^{f}\big)M_{i}^{g}\Delta x_{i} + \sum\limits_{i = 1}^{n}\big(M_{i}^{g} - m_{i}^{g}\big)m_{i}^{f}\Delta x_{i} \le \\ \le A\underbrace{\sum\limits_{i = 1}^{n}\big(M_{i}^{f} - m_{i}^{f}\big)\Delta x_{i}}_{\le \frac{\varepsilon}{2A}} + B\underbrace{\sum\limits_{i = 1}^{n}\big(M_{i}^{g} - m_{i}^{g}\big)\Delta x_{i}}_{\le \frac{\varepsilon}{2B}} \le \Bigg\{\text{В силу интегр. функций } f \text{ и } g\Bigg\} \le \frac{\varepsilon}{2} + \frac{\varepsilon}{2} = \varepsilon.
$$

5. Если функция интегрируема на отрезке $[a, b]$, то она будет интегрируема и на отрезке $[c, d] \subset [a, b]$. 

Действительно, поскольку при сужении отрезка интегрирования разность верхней и нижней интегральных сумм может только уменьшаться.


6. Если функция интегрируема на отрезках $[a, c]$ и $[c, b]$, где $c \in (a, b)$, то она будет интегрируема на отрезке $[a, b]$  

$$
\int\limits_{a}^{b}f(x)dx = \int\limits_{a}^{c}f(x)dx + \int\limits_{c}^{b}f(x)dx, \quad c \in (a, b).
$$

Поскольку функция интегрируема на отрезках $[a, c]$ и $[c, b]$, то найдется такое разбиение этих отрезков, что разность верхней и нижней сумм будет меньше величины $\dfrac{\varepsilon}{2}$ для каждого из них. Тогда объединив эти два разбиения мы получим разбиение отрезка $[a, b]$, для которого разность $S - s \le \varepsilon$ и, кроме того, интегральная сумма по $[a, b]$ будет равна сумме интегральных сумм по отрезкам $[a, c]$ и $[c, d]$.

7. Если функция $f(x) \ge 0$ на сегменте $[a, b]$ и интегрируема на этом сегменте, то

$$
\int\limits_{a}^{b}f(x)dx \ge 0.
$$

Действительно, поскольку каждое слагаемое в соответствующей интегральной сумме будет неотрицательно. 

8. Если $f(x)$ интегрируема на $[a, b]$ и $f(x) \ge m$ то

$$
\int\limits_{a}^{b}f(x)dx \ge m(b - a).
$$

Покажем это:

$$
\int\limits_{a}^{b}\big(f(x) - m\big)dx \ge 0 \text{ - это следует из предыдущего свойства}.
$$

Следовательно,

$$
\int\limits_{a}^{b}f(x)dx \ge \int\limits_{a}^{b}mdx = m(b - a).
$$

9. Если функции $f(x)$ и $g(x)$ интегрируемы на сегменте $[a, b]$ и, кроме того, $f(x) \ge g(x)$ всюду на этом сегменте, то 

$$
\int\limits_{a}^{b}f(x)dx \ge \int\limits_{a}^{b}g(x)dx.
$$

Из свойства 3 следует, что 

$$
\int\limits_{a}^{b}\big(f(x)- g(x)\big)dx = \int\limits_{a}^{b}f(x)dx - \int\limits_{a}^{b}g(x)dx.
$$

Поскольку функция $f(x) - g(x) \ge 0$, то из свойства 7 следует

$$
\int\limits_{a}^{b}\big(f(x)- g(x)\big)dx \ge 0.
$$

Следовательно,

$$
\int\limits_{a}^{b}f(x)dx - \int\limits_{a}^{b}g(x)dx \ge 0 \quad \Rightarrow \int\limits_{a}^{b}f(x)dx \ge \int\limits_{a}^{b}g(x)dx.
$$

10. Если функция $f(x)$ интегрируема на сегменте $[a, b]$, то функция $|f(x)|$ также интегрируема этом отрезке и выполняется следующее неравенство:

$$
\Bigg|\int\limits_{a}^{b}f(x)dx\Bigg| \le \int\limits_{a}^{b}|f(x)|dx. 
$$

В начале покажем, что функция модуля будет интегрируемой. Обозначим через $M_{i}, m_{i}$ - точные верхнюю и нижнюю грани исходной функции и через $M_{i}^{'}, m_{i}^{'}$ - точные верхнюю и нижнюю грани модуля. Возможны три случая:

- $$M_{i} \ge m_{i} \ge 0.$$

В этом случае, $M_{i}^{'} = M_{i}, \quad m_{i}^{'} = m_{i}.$


- $$ 0 \ge M_{i} \ge m_{i}$$

В данном случае, $M_{i}^{'} = |m_{i}|$ и $m_{i}^{'} = |M_{i}|$ и получаем

$$
M_{i}^{'} - m_{i}^{'} = |m_{i}| - |M_{i}| = M_{i} - m_{i},
$$

т.е. получаем то же самое.

- $$M_{i} \ge 0 \ge m_{i}$$

в таком случае, получим что

$$
M_{i} - m_{i} = M_{i} + |m_{i}| \ge 
\begin{cases}
M_{i} - |m_{i}| = M_{i}^{'} - m_{i}^{'}, \text{ если } M_{i} \ge |m_{i}|,\\
|m_{i}| - M_{i} = M_{i}^{'} - m_{i}^{'}, \text{ иначе}.
\end{cases}
$$

Таким образом, получили, что 

$$
S - s \ge S^{'} - s^{'},
$$

следовательно, функция модуля интегрируема.

Теперь покажем, что выполняется приведенное выше неравенство. Поскольку

$$
-|f(x)| \le f(x) \le |f(x)| \quad \underbrace{\Rightarrow}_{\text{свойство } 9} \quad -\int_{a}^{b}|f(x)|dx \le \int_{a}^{b}f(x)dx \le \int_{a}^{b}|f(x)|dx.
$$

А из этого следует, что

$$
\Bigg|\int\limits_{a}^{b}f(x)dx\Bigg| \le \int\limits_{a}^{b}|f(x)|dx. 
$$

11. Пусть функции $f(x)$ и $g(x)$ интегрируемы на сегменте $[a, b]$ и $g(x) \ge 0$. Тогда, если $M$ и $m$ - точные верхняя и нижняя грани функции $f(x)$ на $[a, b]$, то

$$
m\int\limits_{a}^{b}g(x)dx \le \int\limits_{a}^{b}f(x)g(x)dx \le M\int\limits_{a}^{b}g(x)dx. 
$$

Из свойства 4 следует, что произведение функций интегрируемо на рассматриваемом отрезке. А из свойства 9 следует, что справедливо указанное неравенство, поскольку для всех точек $x \in [a, b]$ выполнено

$$
mg(x) \le f(x)g(x) \le Mg(x).
$$

В случае, если $g(x) \le 0$, можно получить следующую оценку

$$
M\int\limits_{a}^{b}g(x)dx \le \int\limits_{a}^{b}f(x)g(x)dx \le m\int\limits_{a}^{b}g(x)dx. 
$$

### Формулы среднего значения:

12. $\underline{\text{Первая формула среднего значения}}$

Пусть функция $f(x)$ интегрируема на сегменте $[a, b]$. Тогда, если $M$ и $m$ - точные верхняя и нижняя грани функции $f(x)$ на $[a, b]$, то найдется такое число $\mu: m \le \mu \le M$, что

$$
\int\limits_{a}^{b}f(x)dx = \mu(b - a). 
$$

Если положить $g(x) = 1$, то из свойства 11 следует, что

$$
m(b - a) \le \int\limits_{a}^{b}f(x)dx \le M(b - a).
$$

Отсюда следует, что число $\mu = \frac{1}{b - a}\int\limits_{a}^{b}f(x)dx$ будет принадлежать отрезку $[m, M]$. 

Кроме того, если функция $f(x)$ непрерывна, то найдется точка $\xi: \mu = f(\xi)$ и соответственно мы получим

$$
\int_{a}^{b}f(x)dx = f(\xi)(b - a). 
$$



13. $\underline{\text{Первая формула среднего значения в обобщенном виде}}$

Пусть функции $f(x)$ и $g(x)$ интегрируемы на сегменте $[a, b]$ и $g(x) \ge 0$. Тогда, если $M$ и $m$ - точные верхняя и нижняя грани функции $f(x)$ на $[a, b]$, то то найдется такое число $\mu: m \le \mu \le M$, что

$$
\int\limits_{a}^{b}f(x)g(x)dx = \mu\int\limits_{a}^{b}g(x)dx.
$$

Если $g(x) = 0$, то, очевидно, можно взять любое $\mu$. Если же $g(x) \neq 0$, то воспользовавшись одним из соотношений из свойства 11, получим

$$
m \le \underbrace{\dfrac{\int\limits_{a}^{b}f(x)g(x)dx}{\int\limits_{a}^{b}g(x)dx}}_{\mu} \le M.
$$

Опять же, если функция  $f(x)$ непрерывна, то найдется точка $\xi: \mu = f(\xi)$ и соответственно мы получим

$$
\int\limits_{a}^{b}f(x)g(x)dx = f(\xi)\int\limits_{a}^{b}g(x)dx.
$$


14. $\underline{\text{Вторая формула среднего значения}}$ (Ильин, Позняк стр. 365)

Если на сегменте $[a, b]$ функция $g(x)$ монотонна, а функция $f(x)$ интегрируема, то найдется такое число $\xi \in [a, b]$:

$$
\int\limits_{a}^{b}f(x)g(x)dx = g(a)\int\limits_{a}^{\xi}f(x)dx + g(b)\int\limits_{\xi}^{b}f(x)dx. 
$$

# 2. Методы вычисления интегралов

### Понятие первообразной функции

$\underline{\text{Определение 4.}}$ Функция $F(x)$ называется **первообразной** для функции $f(x)$ на интервале $(a, b)$, если в любой точке $x \in (a, b)$ функция $F(x)$ дифференцируема и имеет производную $F^{'}(x) = f(x)$.

Из определения видно, что первообразная функции является обратной к величине производной этой функции.

$\underline{\text{Утверждение 3:}}$ Если функции $F_{1}(x)$ и $F_{2}(x)$ - первообразные для функции $f(x)$ на интервале $(a, b)$, то всюду на этом интервале

$$
F_{1}(x) - F_{2}(x) = C = const.
$$


$\underline{\text{Примеры:}}$ 

1. Функция $F(x) = \cos{x} + const$ является первообразной для функции $f(x) = -\sin{x}$:

$$
\big(\cos{x} + const\big)^{'} = \big(\cos{x}\big)^{'} + \big(const\big)^{'} = -\sin{x}.
$$


2. Функция $F(x) = \frac{x}{1 - x^{2}} + const$ является первообразной для функции $f(x) = \frac{1 + x^{2}}{(1 - x^{2})^{2}}$:

$$
\Bigg(\frac{x}{1 - x^{2}} + const\Bigg)^{'} = \Bigg(\frac{x}{1 - x^{2}}\Bigg)^{'} + \big(const\big)^{'} = \frac{x^{'}(1 - x^{2}) - x(1 - x^{2})^{'}}{(1 - x^{2})^{2}} = \frac{1 - x^{2} - x(-2x)}{(1 - x^{2})^{2}} = \frac{1 + x^{2}}{(1 - x^{2})^{2}}.
$$

Рассмотрим произвольную функцию $f(x)$ интегрируемую на отрезке $[a, b]$ и точку $x \in [a, b]$. Тогда величина

$$
F(x) = \int\limits_{a}^{x}f(t)dt,
$$

называется **интегралом с переменным верхним пределом**.

$\underline{\text{Утверждение 4:}}$ Любая непрерывная на отрезке $[a, b]$ функция $f(x)$ имеет на этом отрезке первообразную. Одной из них является функция

$$
F(x) = \int\limits_{a}^{x}f(t)dt. 
$$

Чтобы доказать данное утверждение, нужно показать, что функция $F(x)$ дифференцируема и ее производная равна $f(x)$, то есть нужно показать, что

$$
\lim_{\Delta x \to 0}\frac{F(x + \Delta x) - F(x)}{\Delta x} = f(x).
$$

Покажем это

$$
F(x + \Delta x) - F(x) = \int\limits_{a}^{x + \Delta x}f(t)dt - \int\limits_{a}^{x}f(t)dt = \big\{\text{Св-во 5}\big\} = \int\limits_{a}^{x}f(t)dt + \int\limits_{x}^{x + \Delta x}f(t)dt - \int\limits_{a}^{x}f(t)dt =\\= \int\limits_{x}^{x + \Delta x}f(t)dt = \big\{\text{Формула ср. зн-я}\big\} = f(\xi)\Delta x,
$$
где $\xi \in [x, x + \Delta x]$.

Следовательно,

$$
\lim_{\Delta x \to 0}\frac{F(x + \Delta x) - F(x)}{\Delta x} = \lim_{\Delta x \to 0}f(\xi) = f(x).
$$

Таким образом, мы получили, что операция интегрирования является обратной к операции дифференцирования и, кроме того, мы показали, что существует производная от интеграла с переменным верхним пределом, которая равна подыинтегральной функции:

$$
\frac{d}{dx}\Bigg(\int\limits_{a}^{x}f(t)dt\Bigg) = f(x).
$$

Из двух предыдущих утверждений следует, что любая первообразная $F(x)$ непрерывной на отрезке $[a, b]$ функции $f(x)$ имеет вид

$$
F(x) = \int\limits_{a}^{x}f(t)dt + C.
$$

Положив в данной формуле сначала $x = a$, а затем $x = b$ и используя свойство 1 получим

$$
F(a) = C, \quad F(b) = \int\limits_{a}^{b}f(x)dx + C. 
$$

Отсюда получаем

$$
\int\limits_{a}^{b}f(x)dx = F(b) - F(a).
$$

Данная соотношение называется **основной формулой интегрального исчисления** или **формулой Ньютона - Лейбница**.


$\underline{\text{Примеры:}}$ 

1. $$\int\limits_{0}^{\frac{\pi}{2}}\cos{x}dx = \sin{x}\Bigg|_{0}^{\frac{\pi}{2}} = \sin{\frac{\pi}{2}} - \sin{0} = 1 - 0 = 1.$$ 

2. $$\int\limits_{-1}^{1}e^{x}dx = e^{x}\Bigg|_{-1}^{1} = e - \frac{1}{e}.$$

3. $$\int\limits_{0}^{1}\frac{dx}{1 + x^{2}}dx = arctg x\Bigg|_{0}^{1} = \frac{\pi}{4} - 0 = \frac{\pi}{4}.$$

```python
from sympy import *
import numpy as np

z = Symbol('z')

display(integrate(cos(z), (z, 0, np.pi / 2)))

display(integrate(exp(z), (z, -1, 1)))

display(integrate(1/(z**2 + 1), (z, 0, 1)))
```

### Замена переменных под знаком интеграла

Если выполнены следующие условия

- Функция $f(x)$ непрерывна на сегменте $[a, b]$;

- Сегмент $[a, b]$ является множеством значений некоторой функции $x = g(t)$, определенной на сегменте $[\alpha, \beta]$ и имеющий на этом сегменте непрерывную производную;

- $g(\alpha) = a$ и $g(\beta) = b$,

то справедлива следующая формула

$$
\int\limits_{a}^{b}f(x)dx = \int\limits_{\alpha}^{\beta}f(g(t))g^{'}(t)dt. 
$$

Ильин, Позняк стр.356

$\underline{\text{Примеры:}}$ 

1. 
$$
\int\limits_{-1}^{1}\sqrt{1 - x^{2}}dx = 
\left\{
\begin{array}{ccc}
x = \cos{t} = g(t),\\
f(g(t)) = \sqrt{1 - \cos{t}^{2}} = \sin{t},\\
g^{'}(t) = (\cos{t})^{'} = -\sin{t},\\
t \in [0, \pi]
\end{array}
\right\} = \int\limits_{\pi}^{0}\big(-\sin^{2}{t}\big)dt = \int\limits_{0}^{\pi}\frac{1 - \cos{2t}}{2}dt = \\ = \int\limits_{0}^{\pi}\frac{1}{2}dt - \int\limits_{0}^{\pi}\frac{\cos{2t}}{2}dt = \Bigg(\frac{t}{2} - \frac{\sin{2t}}{4}\Bigg)\Bigg|_{0}^{\pi} = \frac{\pi}{2}
$$


$$
\\ \\
$$

2. $$
\int\limits_{1}^{2}\frac{\ln{x}}{x}dx = 
\left\{
\begin{array}{ccc}
x = e^{t} = g(t),\\
f(g(t)) = \frac{\ln{e^{t}}}{e^{t}} = \frac{t}{e^{t}},\\
g^{'}(t) = e^{t},\\
t \in [0, \ln{2}]
\end{array}
\right\} = \int\limits_{0}^{\ln{2}}\frac{t}{e^{t}}e^{t}dt = \int\limits_{0}^{\ln{2}}tdt = \frac{t^{2}}{2}\Bigg|_{0}^{\ln{2}} = \frac{1}{2}\ln^{2}{2}.
$$

```python
display(integrate(sqrt(1 - z**2), (z, -1, 1)))

display(integrate(ln(z)/ z, (z, 1, 2)))
```

### Интегрирование по частям

Если функции $u(x)$ и $v(x)$ имеют непрерывные производные на сегменте $[a, b]$, то справедлива следующая формула

$$
\int\limits_{a}^{b}u(x)v^{'}(x)dx = \big(u(x)v(x)\big)\Bigg|_{a}^{b} - \int\limits_{a}^{b}v(x)u^{'}(x)dx.
$$


$\underline{\text{Примеры:}}$ 

1. 


$$
\int\limits_{0}^{1}(x + 2)e^{3x}dx =  
\left\{
\begin{array}{ccc}
u = x + 2 \quad \Rightarrow \quad u^{'} = 1,\\
v^{'} = e^{3x} \quad \Rightarrow \quad v = \dfrac{e^{3x}}{3}
\end{array}
\right\} = \\
= (x + 2)\frac{e^{3x}}{3}\Bigg|_{0}^{1} - \int\limits_{0}^{1}\frac{e^{3x}}{3}dx = e^{3} - \frac{2}{3} - \frac{e^{3x}}{9}\Bigg|_{0}^{1} = e^{3} - \frac{2}{3} - \frac{e^{3}}{9} + \frac{1}{9} = \frac{8e^{3}}{9} - \frac{5}{9}.
$$

2.


$$
\int\limits_{1}^{2}x\ln{x}dx =  
\left\{
\begin{array}{ccc}
u = \ln{x} \quad \Rightarrow \quad u^{'} = \frac{1}{x},\\
v^{'} = x \quad \Rightarrow \quad v = \dfrac{x^{2}}{2}
\end{array}
\right\} = \\
= \frac{x^{2}}{2}\ln{x}\Bigg|_{1}^{2} - \int\limits_{1}^{2}\frac{x}{2}dx = 2\ln{2} - \frac{x^{2}}{4}\Bigg|_{1}^{2} = 2\ln{2} - 1 + \frac{1}{4} = 2\ln{2} - \frac{3}{4}. 
$$

```python
display(integrate((z + 2)*exp(3*z), (z, 0, 1)))

display(integrate(ln(z)*z, (z, 1, 2)))
```

# 3. Численные методы интегрирования

Формулу Ньютона - Лейбница и методы замены переменной или интегрирования по частям удобно применять в случае, если первообразная подыинтегральной функции представляет собой некоторую элементарную функцию. В противном случае, необходимо пользоваться методами приближенного вычисления интегралов.  

Давайте рассмотрим соответствующие численные методы. Они основаны на приближении интеграла конечной суммой:


$$
\int\limits_{a}^{b}f(x)dx \approx \sum\limits_{i = 0}^{n}c_{i}f(x_{i}),
$$
 
называется **квадратурной формулой**, числа $c_{i}$ - **коэффициентами квадратурной формулы**, а точки $x_{i} \in [a, b]$ - **узлами**.

Разность

$$
\psi = \int\limits_{a}^{b}f(x)dx - \sum\limits_{i = 0}^{n}c_{i}f(x_{i}),
$$

называется **погрешностью квадратурной формулы**.

Для вычисления квадратурной формулы отрезок $[a, b]$ разбивается на $n$ частичных сегментов с помощью точек $x_{0}, x_{1}, \ldots, x_{n}$:

$$
a = x_{0} < x_{1} < \ldots < x_{n} = b.
$$

Далее, исходный интеграл $\int\limits_{a}^{b}f(x)dx$ представляется в виде суммы интегралов по частичным отрезкам:

$$
\int\limits_{a}^{b}f(x)dx = \sum\limits_{i = 1}^{n}\int\limits_{x_{i - 1}}^{x_{i}}f(x)dx,
$$

и на каждом из этих отрезков $[x_{i - 1}, x_{i}]$ интеграл приближается величиной следующего вида:

$$
\int\limits_{x_{i - 1}}^{x_{i}}f(x)dx \approx g\big(f(x)\big)\alpha\Delta x_{i}, \quad x \in [x_{i - 1}, x_{i}], \quad \Delta x_{i} = x_{i} - x_{i - 1}, \quad \alpha \in \mathbb R.
$$

### Формула прямоугольников

Различают 3 вида формулы прямоугольников:

- *Формула средних прямоугольников:* $g = f\Big(\frac{x_{i - 1} + x_{i}}{2}\Big), \alpha = 1.$


- *Формула левых прямоугольников:* $g = f(x_{i - 1}), \alpha = 1.$


- *Формула правых прямоугольников:* $g = f(x_{i}), \alpha = 1.$


Для каждого из этих случае получам, что

$$
\int\limits_{a}^{b}f(x)dx \approx \sum\limits_{i = 1}^{n}f\Big(\frac{x_{i - 1} + x_{i}}{2}\Big)(x_{i} - x_{i - 1}), \quad \text{ - метод средних прямоугольников,}\\
\int\limits_{a}^{b}f(x)dx \approx \sum\limits_{i = 1}^{n}f(x_{i - 1})(x_{i} - x_{i - 1}), \quad \text{ - метод левых прямоугольников,}\\
\int\limits_{a}^{b}f(x)dx \approx \sum\limits_{i = 1}^{n}f(x_{i})(x_{i} - x_{i - 1}), \quad \text{ - метод правых прямоугольников.}
$$



$\underline{\text{Погрешность методов:}}$ (Самарский, Гулин стр. 162)

$$
\psi \le \frac{\max_{\xi \in [a, b]}f^{''}(\xi)}{24}(b - a)^{3}, \quad \text{ в случае средних прямоугольников}.
$$

$$
\psi \le \frac{\max_{\xi \in [a, b]}f^{'}(\xi)}{2}(b - a)^{2}, \quad \text{ в случае односторонних прямоугольников}.
$$



Реализуем методы прямоугольников и на примере функции

$$
f(x) = 2x + x^{2},
$$

сравним их результат с истинным значением.

```python
import numpy as np

def rectangle_integrate(f, a, b, n, method = 'average'):
    
    x0 = np.array([a + (b - a) / n * i for i in range(n)])
    x1 = np.array([a + (b - a) / n * (i + 1) for i in range(n)])
    
    if (method == 'left'):
        x = x0
    elif (method == 'right'):
        x = x1
    else:
        x = (x0 + x1) / 2

    return sum(f(x) * (np.array(x1) - np.array(x0)))
```

```python
# Сравним результат работы реализованного метода с истинным значением

from sympy import *

a = -2
b = 10
n = 100

print('Rectangle Left  = ', rectangle_integrate(lambda x: 2 * x + x**2, a, b, n, 'left'))
print('Rectangle Right = ', rectangle_integrate(lambda x: 2 * x + x**2, a, b, n, 'right'))
print('Rectangle       = ', rectangle_integrate(lambda x: 2 * x + x**2, a, b, n))

z = Symbol('z')
print('True value      = ', integrate(2*z + z**2, (z, a, b)).evalf())
```

### Формула трапеций

В случае формулы трапеций имеем

$$
g = \frac{f(x_{i - 1}) + f(x_{i})}{2}, \alpha = 1.
$$


Соответствующая оценка для интеграла будет следующей

$$
\int\limits_{a}^{b}f(x)dx \approx \sum\limits_{i = 1}^{n}\frac{f(x_{i - 1}) + f(x_{i})}{2}(x_{i} - x_{i - 1}).
$$



$\underline{\text{Погрешность метода:}}$ (Самарский, Гулин стр. 164)

$$
\psi \le \frac{\max_{\xi \in [a, b]}f^{''}(\xi)}{12}(b - a)^{3}.
$$

Реализуем метод трапеций и на примере той же самой функции

$$
f(x) = 2x + x^{2},
$$

сравним его результат с истинным значением.

```python
def trapezoid_integrate(f, a, b, n):
    
    x0 = np.array([a + (b - a) / n * i for i in range(n)])
    x1 = np.array([a + (b - a) / n * (i + 1) for i in range(n)])

    return sum((f(x0) + f(x1)) / 2 * (np.array(x1) - np.array(x0)))
```

```python
# Сравним результат работы реализованного метода с истинным значением

print('Trapezoid  = ', trapezoid_integrate(lambda x: 2 * x + x**2, a, b, n))
print('True value = ', integrate(2*z + z**2, (z, a, b)).evalf())
```

### Формула Симпсона

В случае формулы Симпсона (парабол) имеем

$$
g = f(x_{i - 1}) + 4f\Bigg(\frac{x_{i - 1} + x_{i}}{2}\Bigg) + f(x_{i}), \alpha = \frac{1}{6}.
$$


Соответствующая оценка для интеграла будет следующей

$$
\int\limits_{a}^{b}f(x)dx \approx \sum\limits_{i = 1}^{n}\Bigg[f(x_{i - 1}) + 4f\Bigg(\frac{x_{i - 1} + x_{i}}{2}\Bigg) + f(x_{i})\Bigg]\frac{(x_{i} - x_{i - 1})}{6}.
$$



$\underline{\text{Погрешность метода:}}$ (Самарский, Гулин стр. 164)

$$
\psi \le \frac{\max_{\xi \in [a, b]}f^{(4)}(\xi)}{2880}(b - a)^{5}.
$$

Реализуем метод Симпсона и на примере той же самой функции

$$
f(x) = 2x + x^{2},
$$

сравним его результат с истинным значением.

```python
def simpson_integrate(f, a, b, n):
    
    x0 = np.array([a + (b - a) / n * i for i in range(n)])
    x1 = np.array([a + (b - a) / n * (i + 1) for i in range(n)])

    return sum((f(x0) + 4 * f((x0 + x1) / 2) + f(x1)) * (np.array(x1) - np.array(x0)) / 6)
```

```python
# Сравним результат работы реализованного метода с истинным значением

print('Simpson    = ', simpson_integrate(lambda x: 2 * x + x**2, a, b, n))
print('True value = ', integrate(2*z + z**2, (z, a, b)).evalf())
```

### Метод Монте - Карло

При приближении значения интеграла методом Монте - Карло рассматривают случайные величины $u_{i}$, имеющие равномерное распределение на отрезке $[x_{i - 1}, x_{i}]:$ $u_{i} \sim R[x_{i - 1}, x_{i}]$.

В этом случае, величина $f(u_{i})$ так же будет случайной величиной, математическое ожидание которой определяется формулой

$$
\mathbb Ef(u_{i}) = \frac{1}{x_{i} - x_{i - 1}}\int\limits_{x_{i - 1}}^{x_{i}}f(x)dx.
$$

Отсюда следует, что

$$
\int\limits_{x_{i - 1}}^{x_{i}}f(x)dx = (x_{i} - x_{i - 1})\mathbb Ef(u_{i}).
$$

А математическое ожидание случайной величины $u_{i}$ можно оценить средним выборочным:

$$
\mathbb Ef(u_{i}) \approx \frac{1}{k}\sum\limits_{j = 1}^{k}f\big(u_{i}^{j}\big).
$$

Тогда оценка для исходного интеграла на отрезке $[a, b]$ будет выглядить следующим образом:

$$
\int\limits_{a}^{b}f(x)dx \approx \sum\limits_{i = 1}^{n}\Bigg(\frac{1}{k}\sum\limits_{j = 1}^{k}f\big(u_{i}^{j}\big)\Bigg).
$$

Таким образом, алгоритм будет следующим:

1. Разбиваем $[a, b]$ на $n$ частичных сегментов;


2. На каждом частичном сегменте генерируем $k$ случайных величин $f(u_{i})$ таких, что случайная величина $u_{i}$ равномерно распределена на отрезке $[x_{i - 1}, x_{i}]$;


3. После этого оцениваем значение интеграла на $i$ - ом частичном сегменте:

$$
\int\limits_{x_{i - 1}}^{x_{i}}f(x)dx \approx \frac{1}{k}\sum\limits_{j = 1}^{k}f\big(u_{i}^{j}\big);
$$


4. Суммируем полученные оценки и получаем оценку для исходного интеграла.

```python
import random

def monte_carlo_integrate(f, a, b, n, k):

    x0 = np.array([a + (b - a) / n * i for i in range(n)])
    x1 = np.array([a + (b - a) / n * (i + 1) for i in range(n)])
    
    s = 0
    for i, j in zip(x0, x1):
        x = np.random.uniform(i, j, k)
        s += sum(f(x) / k * (j - i))
        
    return s
```

```python
print('Monte - Carlo (1)      = ', monte_carlo_integrate(lambda x: 2 * x + x**2, a, b, 1, n))
print('Monte - Carlo (5)      = ', monte_carlo_integrate(lambda x: 2 * x + x**2, a, b, 5, n))
print('Monte - Carlo (10)     = ', monte_carlo_integrate(lambda x: 2 * x + x**2, a, b, 10, n))
print('Monte - Carlo (30)     = ', monte_carlo_integrate(lambda x: 2 * x + x**2, a, b, 30, n))
print('Monte - Carlo (100)    = ', monte_carlo_integrate(lambda x: 2 * x + x**2, a, b, 100, n))
print('True value             = ', integrate(2*z + z**2, (z, a, b)).evalf())
```

```python
import matplotlib.pyplot as plt

a = 0
b = 10
n0 = 10
n1 = 201

rectLeftError = []; rectRightError = []; rectError = []
trapError = []; simpsonError = []; mcError = []

true_int = integrate(cos(z) + z**2, (z, a, b)).evalf()

for i in range(n0, n1):
    r_int_l = rectangle_integrate(lambda x: np.cos(x) + x**2, a, b, i, 'left')
    r_int_r = rectangle_integrate(lambda x: np.cos(x) + x**2, a, b, i, 'right')
    r_int = rectangle_integrate(lambda x: np.cos(x) + x**2, a, b, i)
    t_int = trapezoid_integrate(lambda x: np.cos(x) + x**2, a, b, i) 
    s_int = simpson_integrate(lambda x: np.cos(x) + x**2, a, b, i)
    mc_int = monte_carlo_integrate(lambda x: np.cos(x) + x**2, a, b, i, i)
    
    rectLeftError.append(true_int - r_int_l)
    rectRightError.append(true_int - r_int_r)
    rectError.append(true_int - r_int)
    trapError.append(true_int - t_int)
    simpsonError.append(true_int - s_int)
    mcError.append(true_int - mc_int)
    
    
fig, axs = plt.subplots(figsize = (15, 8), nrows = 1, ncols = 1)

axs.plot(range(n0, n1), rectLeftError, '--', color = 'blue')
axs.plot(range(n0, n1), rectRightError, '.-.', color = 'blue')
axs.plot(range(n0, n1), rectError, color = 'blue')
axs.plot(range(n0, n1), trapError, color = 'green')
axs.plot(range(n0, n1), simpsonError, color = 'black')
axs.plot(range(n0, n1), mcError, color = 'red')

axs.set_xlabel('Количество узлов')
axs.set_ylabel('Ошибка')

axs.legend(['Rectangle Left', 'Rectangle Right', 'Rectangle',
            'Trapezoid', 'Simpson', 'Monte - Carlo'], loc = 'best')
plt.show()
```
