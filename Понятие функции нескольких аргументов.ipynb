# 1. Понятие функции многих переменных


$\underline{\text{Определение 1.}}$ $N$ - мерное пространство называется **евклидовым**, если между любыми двумя его точками $M_{1}$ и $M_{2}$ определено расстояние $\rho(M_{1}, M_{2})$: 

$$
\rho(M_{1}, M_{2}) = \sqrt{(x_{1}^{'} - x_{1}^{''})^{2} + (x_{2}^{'} - x_{2}^{''})^{2} + \ldots + (x_{N}^{'} - x_{N}^{''})^{2}},
$$

которое удовлетворяет следующим условиям:

- **Симметричность**

$$
\rho(M_{1}, M_{2}) = \rho(M_{2}, M_{1});
$$

- **Положительная определенность**

$$
\rho(M_{1}, M_{2}) \ge 0 \quad \forall M_{1} \neq M_{2} \text{ и } \rho(M_{1}, M_{2}) = 0 \quad \forall M_{1} = M_{2};
$$

- **Неравенство треугольника**

$$
\rho(M_{1}, M_{3}) \le \rho(M_{1}, M_{2}) + \rho(M_{2}, M_{3}) \quad \forall M_{1}, M_{2}, M_{3}.
$$


Докажем, что введенная нами функция расстояния удовлетворяет последнему условию:

$$
\begin{multline}
\Big(\rho(M_{1}, M_{3})\Big)^{2} = \sum\limits_{i = 1}^{n}\big(x_{i}^{'} - x_{i}^{'''}\big)^{2} = \sum\limits_{i = 1}^{n}\Big(\big(x_{i}^{'} - x_{i}^{''}\big) + \big(x_{i}^{''} - x_{i}^{'''}\big)\Big)^{2} = \\ = \sum\limits_{i = 1}^{n}\big(x_{i}^{'} - x_{i}^{''}\big)^{2} +2 \sum\limits_{i = 1}^{n}\big(x_{i}^{'} - x_{i}^{''}\big)\big(x_{i}^{''} - x_{i}^{'''}\big) + \sum\limits_{i = 1}^{n}\big(x_{i}^{''} - x_{i}^{'''}\big)^{2} \le \Big\{\text{Нер-во Коши - Буняковского}\Big\} \le \\ \le \underbrace{\sum\limits_{i = 1}^{n}\big(x_{i}^{'} - x_{i}^{''}\big)^{2}}_{\Big(\rho(M_{1}, M_{2})\Big)^{2}} + \underbrace{2\sqrt{\sum\limits_{i = 1}^{n}\big(x_{i}^{'} - x_{i}^{''}\big)^{2}}\sqrt{\sum\limits_{i = 1}^{n}\big(x_{i}^{''} - x_{i}^{'''}\big)^{2}}}_{2\rho(M_{1}, M_{2})\rho(M_{2}, M_{3})} + \underbrace{\sum\limits_{i = 1}^{n}\big(x_{i}^{''} - x_{i}^{'''}\big)^{2}}_{\Big(\rho(M_{2}, M_{3})\Big)^{2}} \le \Big(\rho(M_{1}, M_{2}) + \rho(M_{2}, M_{3})\Big)^{2}
\end{multline}
$$

$\underline{\text{Примером}}$ евклидовых пространств являются множество всех вещественных чисел $\big(\mathbb{R}^{1}\big)$, множество всех пар вещественных чисел $\big(\mathbb{R}^{2}\big)$, троек $\big(\mathbb{R}^{3}\big)$ и т.д.

Найдем расстояние между точками $M_{1} = (0, 1, 1)$ и $M_{2} = (0, -2, 5)$ трехмерного пространства:

$$
\rho(M_{1}, M_{2}) = \sqrt{(0 - 0)^{2} + (1 - (-2))^{2} + (1 - 5)^{2}} = \sqrt{0 + 9 + 16} = 5. 
$$

```python
import numpy as np

x1 = np.array([0, 1, 1])
x2 = np.array([0, -2, 5])
r = np.linalg.norm(x1 - x2)

print(r)
```

$\underline{\text{Определение 2.}}$ Рассмотрим евклидово множество $\mathbf{X} \subset \mathbb{R}^{N}$. Если каждой точке ${\mathbf x} = (x_{1}, x_{2}, \ldots, x_{n}) \in \mathbf{X}$ ставится в соответствие по некоторому известному закону число $y$, причем единственное, то говорят, что на множестве $\mathbf{X}$ задана **функция** $y = f(x_{1}, x_{2}, \ldots, x_{n}) = f({\mathbf x})$. При этом множество $\mathbf{X}$ называется **областью определения** функции $y = f({\mathbf x})$, а множество $\mathbf{Y} = \Big\{y\Big| \exists {\mathbf x} \in \mathbf{X}: f({\mathbf x}) = y\Big\}$ - **множеством значений**. Элемент $y$ из множества значений функции $\mathbf{Y}$ называется **частным значением** функции в точке ${\mathbf x}$. 

$\underline{\text{Примеры:}}$

1. $$y = \sqrt{\sin{\big(x_{1}^{2} + x_{2}^{2}\big)}}$$

Найдем область определения данной функции:

$$
\sin{\big(x_{1}^{2} + x_{2}^{2}\big)} \ge 0 \quad \Rightarrow \quad \big(x_{1}^{2} + x_{2}^{2}\big) \in \big[2\pi k, \pi + 2\pi k\big], \quad k \in {\mathbb N} \cup {0}
$$

Поскольку функция синуса ограничена сверху единицей, то множество значений в данном случае будет отрезком $[0, 1]$.

Построим график области определения:

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize = (8, 8), nrows = 1, ncols = 1)

alpha = np.linspace(0, 2 * np.pi, 200)

for i in range(1, 10):
    i = 10 - i
    radius = np.pi * i
    x1 = radius * np.cos(alpha)
    x2 = radius * np.sin(alpha)
    
    if (i % 2) == 1:
        axs.fill_between(x1, 0, x2, color = 'green', alpha = 0.7)
    else:
        axs.fill_between(x1, 0, x2, color = 'white', alpha = 1)
        
    plt.plot(x1, x2, color = 'black')

axs.set_xlim((-11 * np.pi, 11 * np.pi))
axs.set_ylim((-11 * np.pi, 11 * np.pi))

axs.set_xlabel(r"$x_{1}$", fontsize = 15)
axs.set_ylabel(r"$x_{2}$", fontsize = 15)

axs.set_title(r"Визуализация области определения функции $y = \sqrt{\sin{(x_{1}^{2} + x_{2}^{2})}}$")
plt.show()
```

```python
import pandas as pd

# Зададим сетку
args = np.array(np.meshgrid(np.linspace(-3, 3, 200), np.linspace(-3, 3, 200))).reshape(2, -1)

# Определим функцию и рассчитаем ее значения по сетке
def f(x, y):
    return np.sqrt(np.sin(x**2 + y**2))
    
z = np.vectorize(f)(*args).astype(float)

# Удалим NaN
df = pd.DataFrame(dict(x = args[0], y = args[1], z = z)).dropna()


# Нарисуем
fig = plt.figure(figsize = (25, 20))

# Рисунок 1
ax = fig.add_subplot(1, 2, 1, projection = '3d')
ax.plot_trisurf(df['x'], df['y'], df['z'], cmap='viridis', edgecolor='none')

ax.set_title(r"График функции $y = \sqrt{\sin{(x_{1}^{2} + x_{2}^{2})}}$ на множестве " +\
             r"$[-3; 3] \times [-3, 3]$", fontsize = 15)
ax.set_xlabel(r"$x_{1}$", fontsize = 20)
ax.set_ylabel(r"$x_{2}$", fontsize = 20)
ax.set_zlabel(r"$y$", fontsize = 20)


# Рисунок 2
ax = fig.add_subplot(1, 2, 2, projection = '3d')
ax.plot_trisurf(df['x'][(df.x <= 1.25) & (df.x >= -1.25) & (df.y <= 1.25) & (df.y >= -1.25)], 
                df['y'][(df.x <= 1.25) & (df.x >= -1.25) & (df.y <= 1.25) & (df.y >= -1.25)], 
                df['z'][(df.x <= 1.25) & (df.x >= -1.25) & (df.y <= 1.25) & (df.y >= -1.25)], 
                cmap='viridis', edgecolor='none')

ax.set_title(r"График функции $y = \sqrt{\sin{(x_{1}^{2} + x_{2}^{2})}}$ на множестве " +\
             r"$[-1.25; 1.25] \times [-1.25; 1.25]$", fontsize = 15)
ax.set_xlabel(r"$x_{1}$", fontsize = 20)
ax.set_ylabel(r"$x_{2}$", fontsize = 20)
ax.set_zlabel(r"$y$", fontsize = 20)

plt.show()
```

# 2. Предел функции многих переменных

Рассмотрим функцию $y = f\big(x_{1}, x_{2}, \ldots, x_{n}\big) = f(\mathbf x)$, определенную на некотором множестве $\mathbf{X} \subset \mathbb{R}^{n}$ и точку $M = \big(\tilde{x}_{1}, \tilde{x}_{2}, \ldots, \tilde{x}_{n}\big)$, которая может не принадлежать множеству $\mathbf{X}$, но в любой ее проколотой $\varepsilon$ - окрестности имеются точки из множества $\mathbf{X}$. 


$\underline{\text{Определение 4.}}$ Число $b$ называется **пределом** функции $y = f(\mathbf x)$ в точке $M$, если для любого положительного числа $\varepsilon$ найдется положительное число $\delta$ такое, что для всех точек $\tilde{M}$ из области определения функции, удовлетворяющих условию

$$
0 < \rho\big(\tilde{M}, M\big) < \delta,
$$

выполняется условие

$$
\big|f(\tilde{M}) - b\big| < \varepsilon.
$$


**Отличие предела функции многих переменных от предела функции одной переменной:** основное отличие данных пределов заключается в том, что при нахождении предела функции одной переменной, необходимо исследовать как изменяется значение функции при стремлении к рассматриваемой точке по двум направлением: при движении к данной точке слева по кривой и справа. Соответственно для существования предела необходимо было равенство левого и правого предела. Если же мы имеем дело с функцией нескольких перменных, то таких направлений бесконечно много и для существования предела необходимо, чтобы пределы по всем напрвлениям существовали и были равны.  


$\underline{\text{Примеры:}}$

1. Покажем, что $\lim_{x\to 0, y \to 0}(x + y)\frac{1}{\sin{(x)}}\frac{1}{\cos{(x)}} = 0$:

В соответствии с определением, нам нужно показать, что для любой точки $(x, y)$ такой, что

$$
0 < \sqrt{(x - 0)^{2} + (y - 0)^{2}} = \sqrt{x^{2} + y^{2}} < \delta,
$$

выполняется условие

$$
\big|f(x, y)\big| = \Bigg|(x + y)\frac{1}{\sin{(x)}}\frac{1}{\cos{(x)}}\Bigg| \le \varepsilon.
$$

Из первого условия следует, что 

$$x^{2} + y^{2} \le \delta^{2}.$$ 

Следовательно, 

$$x^{2} \le \delta^{2} \text{ и } y^{2} \le \delta^{2}.$$

Поскольку $\delta > 0$, то

$$
-\delta \le x \le \delta, \quad -\delta \le y \le \delta \quad \Rightarrow \quad |x + y| \le 2\delta.
$$

Учитывая это получим

$$
\Bigg|(x + y)\frac{1}{\sin{(x)}}\frac{1}{\cos{(x)}}\Bigg| \le \Bigg|(x + y)\Bigg|\cdot\Bigg|\frac{1}{\sin{(x)}}\Bigg|\cdot\Bigg|\frac{1}{\cos{(x)}}\Bigg| \le \Bigg|(x + y)\Bigg| \le 2\delta
$$

Таким образом, достаточно взять $\delta = \frac{\varepsilon}{2}$.


2. Покажем, что у функции $f(x, y) = \frac{xy}{x^{2} + y^{2}}$ не существует предела в точке $(0, 0)$. Устремим точку $(x, y)$ к началу координат по прямой $y = kx$:

$$
\lim_{y = kx, x \to 0}f(x, y) = \lim_{x \to 0}\frac{kx^{2}}{x^{2}(1 + k^{2})} = \frac{k}{1 + k^{2}}.
$$

То есть, при движении точки $(x, y)$ к началу координат по различным прямым, получаются разные значения предела. Следовательно, предела в рассматриваемой точке не существует.
